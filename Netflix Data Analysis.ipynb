**Q11: What is the Maximum duration of a Movie or a show ?**

```python
df.duration.unique()
```

```python
#for this q, we will consider minutes only 
df.duration.dtypes
```

```python
# we have to split the duration to get it into ont type
df[['Minutes', 'Unit']] = df['duration'].str.split(' ', expand= True ) #as there is a space b/w duration enteries that's 
                                                                        ##why i have given a space argument in split fn.
```

```python
df.head()
#if we see there is 2 new columns have been created  Minutes and Unit
```

```python
df['Minutes'].dtypes
```

```python
df['Minutes'] = df['Minutes'].fillna(0).astype(int) #changing the datatype of our Minutes column & also giving 0 as dummy value where Minutes= NAN
```

```python
df['Minutes'].max()
```

```python
df[( df['Minutes']== 312)] #verifying our result
```

```python
#Similarly
df['Minutes'].min()
df[( df['Minutes']== 0)] #here 0 implies that dataset doesn't had value and we have given a dummy value zero
```

```python
df['Minutes'].mean()
```
